```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
```

# Data Preparation

```python
file['Date'] = pd.to_datetime(file['Date'])
file['year'] = file['Date'].dt.year
file['month'] = file['Date'].dt.month
```

# correlation with Output column

```python
file.corr()['House_Price_Index'].plot(kind='bar')
```

# House_Price_Index vs Date

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=file['year'], y=file['House_Price_Index'], mode='lines', name='House Price Index'))

fig.update_layout(title='House Price Index Over Years',
                  xaxis=dict(title='Year'),
                  yaxis=dict(title='House Price Index'))

fig.show()
```

```python
fig = px.scatter(file, x='House_Price_Index', y='PPI', title='Scatter Plot: House Price Index vs. PPI')

fig.show()
```
